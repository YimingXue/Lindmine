### 绘制损失曲线图

```python
import scipy.io as sio
import matplotlib.pyplot as plt
```

```python
loss_path = '/home/xueyiming/TEST/snapshots/patch_size27/ResNetv2_2019-03-11_21-57-32'
focalloss_path = '/home/xueyiming/TEST/snapshots/patch_size27/ResNetv2_FocalLoss_2019-03-11_21-43-17'
```

```python
loss = sio.loadmat(loss_path + '/ResNetv2.mat')['loss'][0]
focalloss = sio.loadmat(focalloss_path + '/ResNetv2_FocalLoss.mat')['loss'][0]
```

```python
x = [i for i in range(1250)]
plt.xlim(-1, 50) # 限定横轴范围
# plt.ylim(-1, 100) # 限定纵轴范围
plt.plot(x, focalloss, marker='.', mec='r', mfc='w')
plt.plot(x, loss, marker='.', mec='b', mfc='w')
# plt.legend() # 让图例生效
plt.xlabel("X label") # X轴标签
plt.ylabel("Y label") # Y轴标签
plt.title("A simple plot") # 标题
```

### 绘制函数图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = np.arange(0.001,1,0.001)
ce = -np.log(p)
fl1 = -((1-p)**0.5)*np.log(p)
fl2 = -((1-p)**1)*np.log(p)
fl3 = -((1-p)**2)*np.log(p)
fl4 = -((1-p)**5)*np.log(p)
```

```python
plt.plot(p,ce,color='blue',linewidth=1.5)
plt.plot(p,fl1,color='red',linewidth=1.5)
plt.plot(p,fl2,color='y',linewidth=1.5)
plt.plot(p,fl3,color='m',linewidth=1.5)
plt.plot(p,fl4,color='green',linewidth=1.5)
plt.xlim(0,1)
plt.ylim(0,5)

foo_fig = plt.gcf()
foo_fig.savefig('FocalLossCurve.eps',format='eps',dpi=600)
plt.show()
```
